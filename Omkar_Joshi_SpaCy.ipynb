**Breaking the input text using various methods into:**
* Tokens
* Lemma (Root Word)
* POS
* TAG
* Shape
* is_alpha
* Is the given token a stop word
* Is the given token a punctuation

```python
cols=['Text','Lemma(Root)','POS','TAG','Shape','is_alpha','is_stop', 'is_punct']
sym=[',','.',' ',';',':']
res_sym=[]
rows=[]
for token in doc:
    if token.text not in res_sym and token.text in sym:
        row=token.text, token.lemma_, token.pos_, token.tag_, token.shape_, token.is_alpha, token.is_stop, token.is_punct
        rows.append(row)
        res_sym.append(token.text)
    elif token.text not in sym:
        row=token.text, token.lemma_, token.pos_, token.tag_, token.shape_, token.is_alpha, token.is_stop, token.is_punct
        rows.append(row)
basic_info=pd.DataFrame(rows, columns=cols)
```

```python
pd.set_option('display.max_rows', None)
basic_info
```

**Total Number of Tokens**

```python
print("Total Number of Tokens:", len(basic_info['Text']))
```

**Plotting a dependency graph and Highlighting Named Entities using displacy**

```python
from spacy import displacy

displacy.render(doc, style='dep', jupyter=True, options={'distance':90})
```

```python
displacy.render(doc, style='ent', jupyter=True)
```

```python
for token in doc:
    print(f"{token.text:{12}}{token.morph}")
```
